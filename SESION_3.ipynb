# Unión de bases de datos:

```python
dat1 = {'Nombre':['Carlos','Juan','Sara','Pedro','Luis','Ana'],
        'Edad':[21,23,19,25,19,26]}
dat2 = {'Id':[1029,2516,8437,9289,7373],
        'Nombre':['Ana','Carlos','Sara','Pedro','Luis']}
dat3 = {'Genero':['F','M','M','M','F','M'],
       'Nombre':['Ana','Carlos','Juan','Pedro','Sara','Luis']}
dat4 = {'Nombre':['Carlos','Juan','Sara','Pedro','Luis','Ana'],
       'Programa':['Estadistica','Matematicas','Ing. Sistemas','Ing. Civil','Economia','Ing. Ambiental']}
```

```python
import pandas as pd
df1 = pd.DataFrame(dat1)
df2 = pd.DataFrame(dat2)
df3 = pd.DataFrame(dat3)
df4 = pd.DataFrame(dat4)
```

```python
pd.merge(df1,df2, on = 'Nombre') #Cuando un dato no se encuentra en alguna base de datos no lo considera en el merge
```

```python
pd.merge(df1,df2, on = 'Nombre',how = 'outer') #Para 'Arreglar' este problema ponemos el parametro how = 'outer'
```

```python
datos1 = pd.merge(df1,df2,on= 'Nombre',how = 'outer')
datos1 = pd.merge(datos1,df3, on = 'Nombre',how = 'outer')
datos1 = pd.merge(datos1,df4, on = 'Nombre',how = 'outer')
datos1
```

```python
dfs = [df2,df3,df4]
datos2 = df1
for df in dfs:
    datos2 = pd.merge(datos2,df, on ='Nombre', how='outer')
```

```python
import pandas as pd

bd1 = pd.read_csv('DATOS/DB1.csv')
bd2 = pd.read_csv('DATOS/DB2.csv')
bd3 = pd.read_csv('DATOS/DB3.csv')
bd4 = pd.read_csv('DATOS/DB4.csv')

print(list(bd1))
print(list(bd2))
print(list(bd3))
print(list(bd4))
```

```python
dfs = [bd2,bd3,bd4]
Datos_6 = bd1
for df in dfs:
    Datos_6 = pd.merge(Datos_6,df, on ='Codigo', how = 'outer')
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(10,7), dpi= 90)
plot = Datos_6['Transmision'].value_counts().plot(kind='bar',title='Tipo de transmision')
```

```python
plt.figure(figsize=(10,7), dpi= 90)
plot = Datos_6['Nacionalidad'].value_counts().plot(kind='bar',title='Nacionalidad')
```

```python
plt.figure(figsize=(10,7), dpi= 90)
Datos_61=Datos_6[Datos_6.Nacionalidad=='COL']
resum_61=Datos_61['Marca'].value_counts()
plot = resum_61[resum_61>50].plot(kind='bar',title='Marcas  Colombia')
```
